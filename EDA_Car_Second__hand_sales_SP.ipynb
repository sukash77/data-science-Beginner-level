```python
# Visualize the distribution of Kilometers using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Kilometers'], kde=True, bins=50)
plt.title('Distribution of Kilometers')
plt.xlabel('Kilometers')
plt.ylabel('Frequency')
plt.show()

# Visualize the distribution of Price using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True, bins=50)
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate the sales count for each brand
brand_sales = df['Brand'].value_counts()

# Visualization of highest selling car brand
plt.figure(figsize=(14, 10))

# Get top 10 brands for pie chart
top_10_brands = brand_sales.head(10)
others = brand_sales[10:].sum()

# Create data for pie chart
pie_data = top_10_brands.copy()
pie_data['Others'] = others

# Create pie chart
colors = plt.cm.Set3(np.linspace(0, 1, len(pie_data)))
wedges, texts, autotexts = plt.pie(pie_data.values, labels=pie_data.index, autopct='%1.1f%%',
                                  colors=colors, startangle=90, textprops={'fontsize': 10})

# Make percentages bold
for autotext in autotexts:
    autotext.set_fontweight('bold')
    autotext.set_fontsize(9)

plt.title('Car Brand Market Share (Top 10 + Others)', fontsize=16, fontweight='bold', pad=20)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
